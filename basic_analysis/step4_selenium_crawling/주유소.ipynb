```python
gas_station_zip = gas_station_o[['상호','주소','상표','휘발유']]
gas_station_zip.head()
```

```python
gas_station_zip['휘발유']=gas_station_zip['휘발유'].astype(float)
```

```python
gas_station_sort = gas_station_zip.sort_values('휘발유',ascending=True)
gas_station_zip_low = gas_station_sort[:10]
gas_station_zip_high = gas_station_sort[-10:]
```

```python
from tqdm import tqdm_notebook
```

```python
import requests
def netGetGeocoder(query):    
    url = '''https://dapi.kakao.com/v2/local/search/address.json?query={0}&category_group_code=PO3
    '''.format(query)
    KAKAO_API_KEY = 'f9bcade751569c67742c3c82a2e2f2af'
    headers = {
        'Authorization':'KakaoAK {0}'.format(KAKAO_API_KEY)
    }    
    #API 쿼리 수행
    res = requests.get(url, headers=headers)
    res = res.json()
    if not res['documents']:
        return 0,0
    # 데이터 추출 : 주소, 위도, 경도, 질의어 순으로 튜플 형태로 리턴
    return res['documents'][0]['y'],res['documents'][0]['x']
```

```python
def execute_gps(gas_station):
    addr_lat = []
    addr_lng = []
    for addr in tqdm_notebook(gas_station['주소']):
        tmp = netGetGeocoder(addr)
        addr_lat.append(tmp[0])
        addr_lng.append(tmp[1])
    gas_station['lat']= addr_lat
    gas_station['lng']= addr_lng
    return gas_station
```

```python
gas_station_low=execute_gps(gas_station_zip_low)
gas_station_high=execute_gps(gas_station_zip_high)
```
