```python
import pandas as pd
```

```python
#import sys
#sys.path

p_cctv = pd.read_csv("./data/pusan_cctv.csv",encoding = 'euc-kr')
```

```python
p_cctv.head()
```

```python
p_cctv = p_cctv.iloc[:,2:-1]
```

```python
p_cctv.rename(columns = {'소재지지번주소':'구군'}, inplace = True)
```

```python
p_cctv.head()
```

```python
p_cctv['구군'] = p_cctv['구군'].apply(lambda a: a.split(' ')[1])
```

```python
p_cctv.head()
```

```python
p_cctv = p_cctv.loc[:,['구군','카메라대수']]
p_cctv.head()
```

```python
p_cctv = p_cctv.groupby('구군').sum()
p_cctv.head()
```

```python
p_pop = pd.read_csv("./data/pusan_pop.csv",encoding = 'euc-kr')
p_pop.head()
```

```python
pusan_total = pd.merge(p_cctv, p_pop, on  = '구군')
pusan_total.head()
# inner 똑같음(교집합) left right(각각이 중심이 되서), outer 합집합
```
