```python
# 꼬꼬마 모듈
from konlpy.tag import Kkma
kkma = Kkma()
```

```python
# 미소카츠 (된장소스 커틀릿)「돈까스」 와 된장의 걸작이라 할수있는 
# 「나고야 붉은 된장 소스」가 절묘하게 조화된 미소카츠는 일본 
# 융합문화의 기원지라 할수 있는 나고야의 대표적인 음식으로서 
# 나고야 하면 미소카츠!! 달짝지근한 된장소스를 드셔보시면 의외로 
# 거부감이 없고, 부드러운 맛에 놀라시고 감탄 하시는 분들이 많은 듯 
# 합니다 원래 예전의 포장마차에서 「도테니 (내장을 붉은 된장으로
test_txt ='''

양념한뒤 삶은 요리」의 국물에 쿠시카츠(커틀릿 꼬치)를 찍어 먹던것이 
기원입니다 일본에 오시면 역사깊은 미소카츠를 야바톤에서 꼭 드셔 보세요.
'''
```

```python
# 매개변수
# text : 분석해야될 텍스트
# img : 마스크 이미지
# stopWord : 리스트 형태로 제공해라.(중복해도 상관 없음)
# bgColor : 배경색
# maxWord : 최대글자 캉누트 개수


def wordc_(src_text,src_mask, stopWord, bgcolor='white', maxWord=1000):
    text = open(src_text).read()
    aMask = np.array(Image.open(src_mask))    
    from wordcloud import WordCloud, STOPWORDS
    stopwords = STOPWORDS
    
    stopwords.union(set(stopWord))
    
    wc = WordCloud(background_color=bgcolor,
                   max_words = 1500, 
                   mask = aMask,
                   stopwords = stopwords,
                   margin=12,
                   random_state=1
                  
                  )
    wc1 = wc.generate(text)
    
    
    plt.figure(figsize=(10,10))
    # 이미지를 직접 그리기 
    
    import random
    def color_func(word, font_size, position, orientation, random_state=None, **kwargs):# 함수안에 함수 사용 가능.    
        return 'hsl(178, 235%%, %d%%)'% random.randint(80, 100)
    
    plt.imshow(wc1.recolor(color_func=color_func, random_state = 3), interpolation='bilinear') # ?? interpolation은 뭐냐-보간법
    # 축제거 
    plt.axis('off')
    plt.show()
```

```python
t = Okt() # twitter 분석기
tokens_ko = t.nouns(doc)
print(tokens_ko)
```

```python
ko = nltk.Text(tokens_ko, name = '해외 파병 연장안의 안') # 뭔 의미인지 설명좀
```

```python
# 토큰의 개수
print(len(ko.tokens))
print(len(set(ko.tokens))) # unique 값만 가져옴.
# 단어별 빈도수 
ko.vocab()
```

```python
ko = nltk.Text(tmp,name='해외 파병 연장안의 안') 
plt.figure(figsize=(12,6))
ko.plot(30)
plt.show()
### tip 재귀적 호출은 비동기식으로 작동한다.
```
