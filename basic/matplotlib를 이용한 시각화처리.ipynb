```python
import numpy as np

x = np.arange(0,12,0.01)
x
```

```python
plt.plot(x, np.sin(x))
```

```python
# 격자 배치
plt.plot(x,np.sin(x))
plt.grid()
```

```python
# 라벨 적용
plt.plot(x,np.sin(x))
plt.grid()
# 한글은 일단 깨짐
plt.xlabel('time(시간)')
plt.ylabel('value(값)')
```

```python
# 제목 적용
plt.plot(x,np.sin(x))
plt.grid()
# 한글은 일단 깨짐
plt.xlabel('time(시간)')
plt.ylabel('value(값)')
# 제목
plt.title('sin wave(싸인파)')
plt.show()
```

```python
# 데이터 추가 설정
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.grid()
# 한글은 일단 깨짐
plt.xlabel('time(시간)')
plt.ylabel('value(값)')
# 제목
plt.title('sin wave(싸인파)')
plt.show()
```

```python
# 데이터 범레 추가
plt.figure( figsize=(10,10))
plt.plot(x,np.sin(x),label = 'sin')
plt.plot(x,np.cos(x),label = 'cos')
plt.legend()
plt.grid()
# 한글은 일단 깨짐
plt.xlabel('time(시간)')
plt.ylabel('value(값)')
# 제목
plt.title('sin wave(싸인파)')
plt.show()
```

```python
# 데이터 선 데코레이션(색상)
plt.figure( figsize=(10,10))
plt.plot(x,np.sin(x),label = 'sin', lw = 3 )
plt.plot(x,np.cos(x), 'r',label = 'cos')
plt.legend()
plt.grid()
# 한글은 일단 깨짐
plt.xlabel('time(시간)')
plt.ylabel('value(값)')
# 제목
plt.title('sin wave(싸인파)')
plt.show()
```

```python
# 특정 부분만 발췌
plt.figure( figsize=(10,10))
plt.plot(x,np.sin(x),label = 'sin', lw = 3 )
plt.plot(x,np.cos(x), 'r',label = 'cos')
plt.legend()
plt.grid()
# 한글은 일단 깨짐
plt.xlabel('time(시간)')
plt.ylabel('value(값)')
# 제목
plt.title('sin wave(싸인파)')
# 그래프 중 일부분만 보고 싶다.
plt.xlim((0, np.pi))
plt.ylim(-1, 1.5)
plt.show()
```

```python
# 더미 데이터
x = np.array([1,2,3,2,32,4,5,6,4,33,5,4])
y = np.array([4,3,4,6,7,8,9,5,4,3,21,7])
len(x),len(y)
```

```python
# 점만 표시 => 분포
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker='>')
plt.show()
```

```python
# 데이터 속성을 보여주기 위해서 칼라맵
# 칼라맵은 특정 데이터나 컬럼 값 사용
colormap = x
#c : 칼라맵 데이터
#s : 크기
plt.scatter(x,y,s = 100, c=colormap,marker='>')
plt.show()
```

```python
# 데이터 속성을 보여주기 위해서 칼라맵
# 칼라맵은 특정 데이터나 컬럼 값 사용
colormap = x # x를 기준으로 값에 따라 색깔 조정
#c : 칼라맵 데이터 ()
#s : 크기
plt.scatter(x,y,s = 100, c=colormap,marker='>')
#칼라바
plt.colorbar()
plt.show()
```

```python
# 랜덤 데이터
#loc => 평균
#scale =>확대값(상하)
s1 = np.random.normal(loc=0,scale = 1, size= 1000)
s2 = np.random.normal(loc=5,scale = 0.5, size= 1000)
s3 = np.random.normal(loc=10,scale = 2, size= 1000)
```

```python
plt.plot(s1, label = 's1')
plt.plot(s2,label = 's2')
plt.plot(s3,label = 's3')
plt.legend()
```

```python
# 박스 표시
plt.figure(figsize=(10,6))
plt.boxplot((s1,s2,s3))
plt.grid()
plt.show()
```
