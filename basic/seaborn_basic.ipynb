```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
# 시각화 결과물을 주피터에 내장 시키기 위해 
%matplotlib inline
# 한글처리 
import platform
from matplotlib import font_manager, rc
# 한글처리
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':    # 맥
    rc( 'font', family='AppleGothic' )
elif platform.system() == 'Windows': # 윈도우
    # 폰트 차후 확인
    fontPath = 'c:/Windows/Fonts/malgun.ttf'
    fontName = font_manager.FontProperties( fname=fontPath ).get_name()
    rc( 'font', family=fontName )
else:
    print('알수없는 시스템. 미적용')
```

```python
# 모양
plt.figure(figsize=(10, 6))
# 데이터 설정
x = np.linspace(0, 10, 100)  # 0부터 10사이에 100개의 데이터가 세팅된다. default = 50
y = np.sin(x)
y1 = np.sin(x+0.5) * 2
y2 = np.sin(x+1) * 3
y3 = np.sin(x+1.5) * 4
plt.plot(x, y, x, y1, x, y2, x, y3)
# 데코레이션
# plt.legend()
# 보이기
plt.show()
```

```python
sns.set_style('dark')
# sns.set_style('white')
plt.plot(x, y, x, y1, x, y2, x, y3)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.plot(x, y, x, y1, x, y2, x, y3)
plt.show()
```

### 칼라맵을 사용하는 시본

```python
# 기본데이터를 사용
# tips 라는 데이터를 사용(요일별, 점심/저녁, 흡연여부, 식사금액, 대비 팁)
tips = sns.load_dataset('tips')
tips.head(20)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8,6))
sns.boxplot(x = tips['total_bill'] )
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8,6))
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
# x,y 이외에 추가데이터를 표현
sns.set_style('whitegrid')
plt.figure(figsize=(8,6))
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue ='smoker', palette='Set3')
plt.show()
```

```python
# x,y 이외에 추가데이터를 표현
sns.set_style('whitegrid')
plt.figure(figsize=(8,6))
#무리 형태
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color='.5')
plt.show()
```

```python
# x,y 이외에 추가데이터를 표현
sns.set_style('whitegrid')
plt.figure(figsize=(8,6))
#무리 형태
sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette='Set2')
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color='.25')  # whisker 밖에 있는 점은 표시 안하게 하는 방법은 없나.
plt.show()
```

```python
sns.set_style('darkgrid')
#  데이터를 scatter(산점도)처럼 표현, 직선이 표시되는 형태
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# smoker 추가
sns.set_style('darkgrid')
# 데이터를 scatter(산점도)처럼 표현, 직선이 표시되는 형태
# total_bill 변수와 tip 변수간의 상관관계를 lmplot() 통해서 확인
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, palette='Set1')
plt.show()
```

### 산점도(scatter plot)
- 변수간의 관계성 확인
- 대각선을 기준으로 좌우 대칭적으로 plot이 나오는 산점도 행렬

```python
sns.set(style='ticks')
# 아이리스 데이터 로드
# 꽃잎, 꽃받침 너비와 폭을 가지고 종을 구분
iris = sns.load_dataset('iris')
iris.head()
```

```python
# 'setosa', 'versicolor', 'virginica' => 레이블(label)
iris['species'].unique()
```

```python
# 대각선 정보를 제외한 변수 상호간의 관계성 확인(변수끼리 독립적인가?)
sns.pairplot(iris)
plt.show()
```

```python
# setosa는 변수간에 중첩성이 안보여서 판독이 비교적 명확하다.
# 'versicolor', 'virginica'  이 2개 레이블 데이터는 대부분은 명확하게 변수간 구분이 된다.
# sepal_length<-> sepal_width 이 부분이 중첩 부분이 많아서 다른 파트에서 좀 더 강화하여 체크할 필요가 있다.
# 판독의 가중치를 다른 쪽 대비로 가는게 유리
# 변수 4개의 대비 species 추가된 형태
sns.pairplot(iris, hue='species')
plt.show()
```

```python

sns.pairplot(iris, vars=['sepal_width','sepal_length'],hue='species')
plt.show()
```

```python
# 특정 변수끼리의 데이터만
sns.pairplot(iris, x_vars=['sepal_width','sepal_length'],y_vars=['petal_width','petal_length'], hue='species')
plt.show()
```

```python
anscombe=sns.load_dataset('anscombe')
anscombe.head()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='x', y='y', data=anscombe.query("dataset =='I'"))
plt.show()
```

```python
sns.set_style('darkgrid')
# import, regplot
# scatter_kws : 변곡점의 크기
# order : 차수 지정, np.polyfit - 
sns.lmplot(x='x', y='y', data=anscombe.query("dataset =='II'"), order=2, size=7, scatter_kws={'s':100})
plt.show()
```
