## 데이터 가공 및 전처리

```python
# 데이터 시트(8124 * 22) 그리고 값이 ASCII, 레이블 데이터(Series) 준비 => label
```

```python
label = mr.iloc[:,0] #mr[mr.columns[0]]

datas = mr.iloc[:,1:] # ascci => ord를 이용하면 구할 수 있음.
```

```python
type(datas)
```

```python
datas.shape
```

```python
#### 각 내용을 ascii 코드로 바꾸기

col_nums = datas.shape[1]
row_nums = datas.shape[0]
```

```python
#방법1
start1_time = time.time()
end1_time = time.time()
print(end1_time - start1_time)

def each_element_apply(data_frame,func):
    import pandas as pd         
    dataframe_container = []
    for col in data_frame.columns:
        series_container = []
        for a in data_frame[col]:       
            series_container.append(func(a)) # 안에 돌아가는게 빠른연산.
        dataframe_container.append(series_container) # 20 x 8124
    return pd.DataFrame(dataframe_container).T
```

```python
each_element_apply(datas,ord)
```

```python
#방법1
start1_time = time.time()
dataframe_container = []

for col_num in range(1,col_nums):       #for문이 적게 돌아감.
    series_container = []
    for a in datas[col_num]:       
        series_container.append(ord(a)) # 안에 돌아가는게 빠른연산.
    dataframe_container.append(series_container) # 20 x 8124
pd.DataFrame(dataframe_container).T.head()
end1_time = time.time()
print(end1_time - start1_time)
```

```python
# 방법2
start1_time = time.time()
dataframe_container = []

for row_num in range(row_nums):
    rows_container = []
    for a in datas.iloc[row_num,:]:       # for문이 많이 돌아감.
        rows_container.append(ord(a))
    dataframe_container.append(rows_container) # 20 x 8124
pd.DataFrame(dataframe_container).head()
end1_time = time.time()
print(end1_time - start1_time)
```

```python
# 방법3(이게 제일 깔끔함.)
start1_time = time.time()
data = datas.apply(lambda x : x.apply(lambda n:ord(n)), axis = 1) # series 에다가 apply를 한번 더 적용 할 수 있음.
data.head()
end1_time = time.time()
print(end1_time - start1_time)
```

```python
# dataFrame에서 개별값을 조작할때
start1_time = time.time()
data = datas.applymap(lambda x: ord(x)) # 여기서 x는 값.
data.head()
end1_time = time.time()
print(end1_time - start1_time)
```

```python
# 데이터 확인
print(label.shape, data.shape)
```

```python
# 훈련데이터와 테스트 데이터
data_train, data_test, label_train, label_test = train_test_split(data,label,random_state=1)
print(data_train.shape, data_test.shape, label_train.shape, label_test.shape)
```

## 모델 생성 및 학습

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(data_train, label_train)
```

## 예측

```python
pre = clf.predict(data_test)  
```

```python
metrics.accuracy_score(pre,label_test)
```

```python
acScore = metrics.classification_report(label_test,pre)
print(acScore)
```

```python
#데이터 재구성
max_col=max([len(tmp[n].unique()) for n in tmp.columns])
df = pd.DataFrame()##? 타입이 DataFrame인지 아닌지 어떻게 아는가???
#대상 데이터
mrTmp = mr.loc[:, 1:]
# mrTmp에 컬럼 기준으로 한줄씩 데이터를 뽑아서 12개의 0과 1의 조합으로 구성
# 22번 반복 
for col in mrTmp.columns:
    #컬럼 기준으로 : col        
    size = mrTmp[col].shape  #컬럼 한줄씩 뽑아서 : (8124)
    # 한줄의 데이터에서 특성으로 사용하는 값들을 열거해라.
    
    key = mrTmp[col].unique()
    value = dict(zip(key, range(len(mrTmp[col].unique())))) # zip 한뒤에 dict를 해줘야 비로소 완성
    target = np.zeros((size[0], max_col),dtype=int)
    
    # i 인덱스 , v 값
    for i,v in enumerate(mrTmp[col]):
        # 머릿속에 떠오르지 않는다면 직접 적어보거나, 적기도 힘들다면 ,print() 로 알아 보자.
        # 0번 루프 : 0, x
        # 1번 루프 : 1, x
        # 2번 루프 : 2, x ...
        target[i][value[v]] = 1 
        
        
    # 값세팅
    
    #dataframe에 이어 붙이기 위해서
    df = pd.concat([df,pd.DataFrame(target)],axis =1)    

df.columns = [n for n in range(df.shape[1])]
df.head()
    
    
```

```python
# 훈련데이터와 테스트 데이터
data_train, data_test, label_train, label_test = train_test_split(df,label,random_state=1)
print(data_train.shape, data_test.shape, label_train.shape, label_test.shape)
```

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(data_train, label_train)
```

```python
pre = clf.predict(data_test)
```

```python
metrics.accuracy_score(pre,label_test)
```

```python
acScore = metrics.classification_report(label_test,pre)
print(acScore)
```
