### 머신러닝

**1. 정의**

- 수많은 데이터를 학습시켜서, 여기에서 있는 패턴을 찾는것
- 패턴만 찾아내면,이를 기반으로 **데이터를 분류하거나** **미래를 예측**할 수 있다.

**2. 개요**

- 인공지능 분야의 한 과제
- 인간의 뇌가 자연스럽게 수행하는 '학습'이라는 능력을 컴퓨터로 구현한 방법
- 문자인식, 음성인식, 게임 전략(바둑,장기), 의료진단, 로봇 개발 등 다방면 사용

** 3. 방법 **

- 벡터가 가장 중요!!
- (vector)는 공간에서 크기와 방향을 갖는 것
- 머신러닝은 계산을 데이터간의 구분선을 찾아낼때 사용
- 구분선을 찾을때 사용한 점(데이터)와 점(데이터) 사이에 거리를 구하는데 사용
- 최근접이웃법 예를 들면 특정 데이터와 가까우면 비슷한 데이터라고 판정

**4. 특징(특성: feature 추출)**

- 머신러닝 구동시 데이터가 어떤 특성을 가지고 있는지 벡터로 사용
- 특성 추출이라고 정의
- 예
> 메일을 학습에 데이터로 사용한다. -> 어떤 특성을 사용할 것인가?
> 특정 글자의 빈도, 금지어, 단어출편빈도, 등등
- 어떤 특성을 사용할 것인지 관건!! -> 작업자가 직접 구현하고 머신러닝의 성과에도 영향을 미친다.

**5. 머신러닝 종류**

- 지도 학습(교사학습)
> - 데이터와 답을 함께 입력(입력과 출력이 이미 존재)
> - 새로운 데이터(처음 데이터)를 넣어서 답을 예측
- 비지도 학습(비교사 학습)
> - 데이터는 입력하는데, 답은 입력하지 않는다.(입력만 존재)
> - 새로운 데이터의 규칙성을 찾는데 포인트
- 강화학습
> - 부분적으로 답을 입력
> - 데이터를 기반으로 최적의 답을 찾는 과정

**6. 지도 학습(교사학습)**
- supervised learning
- 데이터 입력시 답(레이블)을 함께 입력한다. 
- 예)글자인식 
> 입력 : 글자가 나타내는 이미지 데이터 
> 결과 : 해당 글자가 무엇을 의미하는지 
> - 위에 2개를 넣어서 학습
> - 학습의 결과로 머신러닝 모델을 구축하면 새로운 글자 이미지를 넣어서 해당 이미지가 어떤 글자인지 예측한다.


**7.비지도학습(비교사 학습)**

- unsupervised learning
- 입력은 있으나 출력이 ㅇㅄ는 형태
- 최정적으로 내야 하는 답이 정해져 있지 않다.(차별성)
- 예)
> - 사람도 기본적으로 제대로 알 수 없는 본질적인 구조들 
> - 클러스턱 분석
> - 주성분 분석
> - 벡터 양자화
> - 자기 조직화

**8. 강화학습**

- Reinforcement learning
- 현재 상태를 관찰 -> 어떻게 대응해야하는지 관련 문제를 다루는 분야
- 지도와 유사하지만 답을 완전하게 제공하진 않는다.
- 구성 : 행동의 주체 + 환경(상황, 상태)
- 예
> - 행동의 주체 : 애완견
> - 환경 : 자동으로 먹이를 주는 기계
> - 1) 기계에서 사료 냄새가 나는 것을 애완견이 발견한다. 
> - 2) 사료를 먹는 방법을 모른다.
> - 3) 기계 주변을 이렇게 저러헥 건드려 본다.(치기,비비기, ... )
> - 4) 우연히 스위치를 눌렀는데 사료가 나옴 -> 행동에 대한 보상을 받음
> - 5) 애완견은 또 여러분 ㅣㅅ도해보고 사료가 나온 것을 깨달았다.
> - 6) 배가 고프면 스위치를 눌러서 사료를 먹을 수 있다라는 명제를 이해
> - 7) 애완견이 스스로 행동을 통해 학습을 한 것이다. -> 이를 컴퓨터로 구현한 것이 강화학습이다.


**9. 머신 러닝 흐름**
- 명제 : 요구사항, 주제
- 데이터 수집 
- 데이터 가공
> 어떤 특성을 사용할 것인가? 어떤 형식을 가공할 것인가?
- 데이터 학습 -> 방법:(여러 알고리즘) -> 매개변수(파라미터) 튜닝 -> 모델학습
- 모델 평가 
> 정밀도(요구사항에 나오는 기준값) 가 나오면 성공 아니면 실패 
> 실패시 다시 반복

**10. 머신러닝 응용 분야 **

- 클래스 분류(classificaiton)
> 특정 데이터에 레이블을 붙여 분류
> 스팸메일 분류, 필기 인식, 증권사기

- 클러스터링 -군집화(clustering)
> 값의 유사성으로 데이터를 여러그룹으로 나눈다.
> 예) 사용자의 취향을 그룹으로 묶어서 광고

- 추천(Recommendation)
> 특정 데이터를 기반으로 다른 데이터를 추천
> 예) 인터넷 서점에서 책을 샀다 - > 다른 책 추천, 넷플릭스에 영화를 봤다 -> 추천, 유튜브 추천

- 회귀(Regression)
> 과거의 데이터로 미래의 데이터를 예측 하는 것
> 예) 판매 예측, 주가 변동

- 차원축소(Dimensionality reduction)
> 데이터의 특성을 유지하면서 데이터의 양을 줄이는 과정
예) 데이터 시각화, 구조 추추릉ㄹ 통한 용량 을 줄여서 계산 속도 업그레이드

- 과적합(Overfitting) -초과학습
> 훈련 데이터가 학습 되어 있다. 그런데 새로운 데이터에 대한 예측을 잘 못한다.
> 일반화가 되어 있지 않다.
> 예) 데이터가 적다면 데이터를 보충, 모델이 비해서 문제가 너무 복잡하다. -> 모델 변경
